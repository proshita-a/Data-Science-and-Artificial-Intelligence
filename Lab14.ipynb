### PART 1:

```python
#step 1
from nltk.tokenize import word_tokenize
para = "Climate change is affecting ecosystems across the planet. Rising temperatures are causing glaciers to melt at unprecedented rates. Scientists are studying long-term patterns to predict future environmental impacts.Communities worldwide are working together to develop sustainable solutions."
tokens = word_tokenize(para)
word_tokens = [t for t in tokens if t.isalpha()]
print("Number of tokens:", len(word_tokens))
```

```python
#step 2
from nltk.tokenize import sent_tokenize
sentence = sent_tokenize(para)
print("Number of sentence tokens: ", len(sentence))
```

```python
#step 3
from nltk.corpus import stopwords

#getting english stopwords
stop_words = set(stopwords.words("english"))

filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print("Original tokens:", tokens)
print("After removing stopwords:", filtered_tokens)
print("Number of tokens after stopword removal:", len(filtered_tokens))
```

```python
#step 4(a)
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed = [stemmer.stem(word) for word in filtered_tokens]

print("Filtered tokens:", filtered_tokens)
print("Final processed words (stemmed):", stemmed)
```

```python
#step 4(b)
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp(para)

filtered = [token for token in doc if not token.is_stop and not token.is_punct]

lemmas = [token.lemma_ for token in filtered]

print("Filtered tokens:", [t.text for t in filtered])
print("Final processed words (lemmatized):", lemmas)
```

```python
#step 5

# Menu Loop
while True:
    print("\n--- TEXT PROCESSING MENU ---")
    print("1. Tokenize the dataset into words")
    print("2. Remove stop-words")
    print("3. Apply stemming")
    print("4. Exit")
    
    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        print("\nWord Tokens:")
        print(tokens)
        print("Number of tokens:", len(tokens))

    elif choice == '2':
        filtered = [w for w in tokens if w.lower() not in stop_words]
        print("\nAfter Removing Stopwords:")
        print(filtered)
        print("Number of words after stopword removal:", len(filtered))

    elif choice == '3':
        filtered = [w for w in tokens if w.lower() not in stop_words]
        stemmed = [stemmer.stem(w) for w in filtered]
        print("\nAfter Stemming:")
        print(stemmed)

    elif choice == '4':
        print("\nExiting program...")
        break

    else:
        print("Invalid choice! Please enter 1–4.")
```

### Why is it important to preprocess text before applying NLP algorithms?
Preprocessing is essential because raw text is messy and inconsistent. Cleaning and normalizing the text helps NLP algorithms work more accurately and efficiently. It removes noise (like punctuation and stopwords), ensures consistency (through lowercasing, stemming, or lemmatization), reduces dimensionality, and extracts meaningful features that models can understand. Overall, preprocessing improves both the quality of input data and the performance of NLP models.

### Part B

```python
# step 1 completed above
# Step 2: basic chatbot Using if-else logic

def basic_chatbot():
    print("Basic Chatbot Activated! Type 'bye' or 'exit' to stop.\n")
    
    while True:
        user = input("You: ").lower()

        if user in ["hello", "hi"]:
            print("Bot: Hello! Nice to meet you.")
        
        elif user == "how are you":
            print("Bot: I’m feeling great and ready to chat!")
        
        elif user == "what is your name":
            print("Bot: I’m ChatBot 1.0, your simple rule-based assistant!")
        
        elif user in ["fun fact", "tell me something fun"]:
            print("Bot: Did you know? Honey never spoils — archaeologists found 3000-year-old honey still edible!")
        
        elif user in ["bye", "exit"]:
            print("Bot: Goodbye! Have a great day!")
            break
        
        else:
            print("Bot: I am sorry, I don’t understand. Can you rephrase?")

# Step 3: Pattern-Matching Chatbot

def pattern_chatbot():
    print("Pattern-Matching Chatbot Activated! Type 'bye' or 'exit' to stop.\n")
    
    while True:
        user = input("You: ").lower()

        if "hello" in user or "hi" in user:
            print("Bot: Hello! How can I help you today?")
        
        elif "how are you" in user:
            print("Bot: I’m doing well! Thanks for asking.")
        
        elif "name" in user:
            print("Bot: I'm PatternBot, here to assist you!")
        
        elif "fun" in user or "fact" in user:
            print("Bot: Fun fact: Bananas are berries, but strawberries are not!")
        
        elif "bye" in user or "exit" in user:
            print("Bot: Goodbye! It was nice talking to you.")
            break
        
        else:
            print("Bot: I am sorry, I don’t understand. Can you rephrase?")

basic_chatbot()
pattern_chatbot()
```

### 1) Explain the difference between the basic if-else chatbot and the pattern-matching chatbot.
The basic if-else chatbot only responds when the user enters an exact matching phrase.
For example, it understands "hello" but not "hello there".

The pattern-matching chatbot looks for keywords inside the sentence, so it can understand more flexible inputs like "hi bot", "tell me a fun fact", or "my name is...".

It is more natural and user-friendly because it doesn't require exact text matching.

### 2) Give an example of a situation where a rule-based chatbot may fail.
A rule-based chatbot may fail when the user inputs something unexpected or complex that does not match any predefined patterns.

Example:
User: “Can climate change affect food production in coastal regions?”
A rule-based bot will not understand this because it requires reasoning, context understanding, or domain knowledge.
