## Part 1:

```python
import pandas as pd
import numpy as np

df = pd.read_csv("customer_dataset.csv")
df.head()
```

```python
# Overall mean of Total Spend
overall_mean = df["Total Spend"].mean()
print("Overall Mean (Total Spend):", overall_mean)
```

```python
# Group by Membership Card type
group_stats = df.groupby("Membership Card")["Total Spend"].agg(["mean", "var", "count"])
print(group_stats)
```

```python
# Calculating SSG
SSG = sum(group_stats["count"] * (group_stats["mean"] - overall_mean)**2)
print("SSG (Between Groups):", SSG)
```

```python
# Calculating SSE
SSE = 0
for name, group in df.groupby("Membership Card"):
    group_mean = group["Total Spend"].mean()
    SSE += sum((group["Total Spend"] - group_mean)**2)

print("SSE (Within Groups):", SSE)
```

```python
# Number of groups (k) and total observations (N)
k = df["Membership Card"].nunique()
N = len(df)

# Degrees of freedom
df_between = k - 1
df_within = N - k

print("df_between:", df_between)
print("df_within:", df_within)
```

```python
#Mean squares
MSG = SSG / df_between
MSE = SSE / df_within

print("MSG (Mean Square Between Groups):", MSG)
print("MSE (Mean Square Within Groups):", MSE)
```

```python
#f stat
F_statistic = MSG / MSE
print("F-statistic:", F_statistic)
```

There are signifiacnt changes the amount spend depending upon membership level

### PART 2:

```python
# grouping dataset by membership card type
groups = [group["Total Spend"].values for name, group in df.groupby("Membership Card")]
for i, g in enumerate(groups, 1):
    print(f"Group {i} sample size: {len(g)}")
```

```python
print("Null Hypothesis (H0): There are no significant differences in spending among the membership types.")
print("Alternative Hypothesis (H1): There are significant differences in spending among the membership types.")
```

```python
from scipy.stats import f_oneway

# One-way ANOVA
anova_result = f_oneway(*groups)
print(anova_result)
```

in the above code, groups is actually a list of arrays so i used * which automatically unpacks the list so each array is passed as its own argument. asterisk here is an unpacking operator

```python
F_statistic = anova_result.statistic
p_value = anova_result.pvalue

print("F-statistic:", F_statistic)
print("p-value:", p_value)
```

```python
alpha = 0.05  # significance level
if p_value < alpha:
    print(f"Since p-value ({p_value:.4f}) < 0.05, we reject H0.")
    print("Conclusion: There are significant differences in spending among the membership types.")
else:
    print(f"Since p-value ({p_value:.4f}) >= 0.05, we fail to reject H0.")
    print("Conclusion: There are no significant differences in spending among the membership types.")
```

In this lab, we analyzed customer spending data across different membership card types using one-way ANOVA.

Part 1 (Manual Computation):
We calculated the overall mean, group-wise means and variances, and then computed SSG (between groups) and SSE (within groups). From these, we derived the MSG, MSE, and F-statistic. A relatively high F-statistic value suggests potential differences in average spending across membership types.

Part 2 (SciPy ANOVA):
Using scipy.stats.f_oneway, we obtained the F-statistic and p-value directly.

Null Hypothesis (H0): No significant difference in spending among membership card types.

Alternative Hypothesis (H1): At least one group differs significantly.

Interpretation:
Based on the p-value at α = 0.05, we decide whether to reject H0.

If p < 0.05 → reject H0 → significant differences exist.

If p ≥ 0.05 → fail to reject H0 → no significant differences found.

Thus, the lab demonstrates both the manual calculation process of ANOVA and its practical application using statistical libraries.
