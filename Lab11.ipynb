### QUESTION 1

```python
# Breadth First Search

n = int(input("Enter the number of rows: "))
m = int(input("Enter the number of columns: "))

grid = []
for i in range(n):
    row = []
    for j in range(m):
        element = int(input(f"Enter element for grid[{i}][{j}]: "))
        row.append(element)
    grid.append(row)

start = (0, 0)
goal = (n - 1, m - 1)
posn = start

def up(posn):
    (posn[0] - 1, posn[1])

def down(posn):
    return (posn[0] + 1, posn[1])

def right(posn):
    return (posn[0], posn[1] + 1)

def left(posn):
    return (posn[0], posn[1] - 1)

visited = [posn]

def bfs(grid, posn, visited):
    n = len(grid)
    m = len(grid[0])

    if (posn[0] + 1 < n) and (grid[posn[0] + 1][posn[1]] == 0) and (down(posn) not in visited):
        new_posn = down(posn)
        visited.append(new_posn)
        return new_posn

    elif (posn[1] + 1 < m) and (grid[posn[0]][posn[1] + 1] == 0) and (right(posn) not in visited):
        new_posn = right(posn)
        visited.append(new_posn)
        return new_posn

    elif (posn[1] - 1 >= 0) and (grid[posn[0]][posn[1] - 1] == 0) and (left(posn) not in visited):
        new_posn = left(posn)
        visited.append(new_posn)
        return new_posn

    elif (posn[0] - 1 >= 0) and (grid[posn[0] - 1][posn[1]] == 0) and (up(posn) not in visited):
        new_posn = up(posn)
        visited.append(new_posn)
        return new_posn

    else:
        return -1


while posn != goal:
    new_pos = bfs(grid, posn, visited)
    print("Moved to:", new_pos)
    if new_pos == -1:
        print("No solution exists")
        break
    posn = new_pos

print("Number of steps taken:", len(visited)-1)
```

### QUESTION 2

```python
# Depth First Search (DFS)

def dfs_all_paths(graph, start, end):
    paths = []
    stack = [(start, [start])]

    while stack:
        node, path = stack.pop()
        if node == end:
            paths.append(path)
        else:
            for neighbor in sorted(graph[node], reverse=True):
                if neighbor not in path:
                    stack.append((neighbor, path + [neighbor]))

    paths.sort()
    return paths

n = int(input("Enter the number of nodes: "))
graph = {}
for i in range(n):
    neighbors = list(map(int, input(f"Enter neighbors of node {i} (space-separated): ").split()))
    graph[i] = neighbors

start = int(input("Enter start node: "))
end = int(input("Enter end node: "))

result = dfs_all_paths(graph, start, end)
print("All paths:", result)
```

### QUESTION 3

```python
# Uniform Cost Search

def ucs_shortest_path(graph, start, goal):
    queue = [(start, [start], 0)]
    visited = {}

    while queue:
        min_index = 0
        min_cost = float('inf')
        for i, (node, path, cost) in enumerate(queue):
            if cost < min_cost:
                min_cost = cost
                min_index = i

        node, path, cost = queue.pop(min_index)

        if node == goal:
            return (path, cost)

        if node in visited and visited[node] <= cost:
            continue
        visited[node] = cost

        for (neighbor, edge_cost) in graph[node]:
            new_cost = cost + edge_cost
            queue.append((neighbor, path + [neighbor], new_cost))

    return (None, 0)

n = int(input("Enter number of nodes: "))
graph = {}
for i in range(n):
    print(f"Enter neighbors and costs for node {i} (format: neighbor cost ...): ")
    data = input().split()
    pairs = []
    for j in range(0, len(data), 2):
        neighbor = int(data[j])
        cost = int(data[j+1])
        pairs.append((neighbor, cost))
    graph[i] = pairs

start = int(input("Enter start node: "))
goal = int(input("Enter goal node: "))

path, total_cost = ucs_shortest_path(graph, start, goal)
if path:
    print("Path found:", path)
    print("Total cost:", total_cost)
else:
    print("No path found. Cost = 0")
```

### QUESTION 4

```python
# A* Search

def manhattan(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

def astar_shortest_path(grid):
    n, m = len(grid), len(grid[0])
    start = (0, 0)
    goal = (n - 1, m - 1)
    open_list = [(start, [start], 0)]
    closed = []

    while open_list:
        best_index = 0
        best_f = float('inf')
        for i, (pos, path, g) in enumerate(open_list):
            f = g + manhattan(pos, goal)
            if f < best_f:
                best_f = f
                best_index = i

        pos, path, g = open_list.pop(best_index)
        closed.append(pos)

        if pos == goal:
            return (path, g)

        for dr, dc in [(1,0), (-1,0), (0,1), (0,-1)]:
            nr, nc = pos[0] + dr, pos[1] + dc
            if 0 <= nr < n and 0 <= nc < m and grid[nr][nc] == 0:
                new_pos = (nr, nc)
                if new_pos not in closed:
                    open_list.append((new_pos, path + [new_pos], g + 1))

    return (None, -1)

n = int(input("Enter rows: "))
m = int(input("Enter cols: "))
grid = [list(map(int, input(f"Enter row {i}: ").split())) for i in range(n)]

path, cost = astar_shortest_path(grid)
print("Path:", path)
print("Cost:", cost)
```

```python
# Best-First Search

def manhattan(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

def best_first_search(grid):
    n, m = len(grid), len(grid[0])
    start = (0, 0)
    goal = (n - 1, m - 1)
    open_list = [(start, [start])]
    visited = []

    while open_list:
        best_index = 0
        best_h = float('inf')
        for i, (pos, path) in enumerate(open_list):
            h = manhattan(pos, goal)
            if h < best_h:
                best_h = h
                best_index = i

        pos, path = open_list.pop(best_index)
        visited.append(pos)

        if pos == goal:
            return (path, len(path) - 1)

        for dr, dc in [(1,0), (-1,0), (0,1), (0,-1)]:
            nr, nc = pos[0] + dr, pos[1] + dc
            if 0 <= nr < n and 0 <= nc < m and grid[nr][nc] == 0:
                new_pos = (nr, nc)
                if new_pos not in visited:
                    open_list.append((new_pos, path + [new_pos]))

    return (None, -1)

n = int(input("Enter rows: "))
m = int(input("Enter cols: "))
grid = [list(map(int, input(f"Enter row {i}: ").split())) for i in range(n)]

path, cost = best_first_search(grid)
print("Path:", path)
print("Steps:", cost)
```

### QUESTION 6

```python
# Greedy Best-First Search

def manhattan(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

def bfs_shortest_steps(grid):
    n, m = len(grid), len(grid[0])
    start = (0, 0)
    goal = (n - 1, m - 1)
    queue = [(start, 0)]
    visited = [start]

    while queue:
        pos, steps = queue.pop(0)
        if pos == goal:
            return steps
        r, c = pos
        for dr, dc in [(1,0), (-1,0), (0,1), (0,-1)]:
            nr, nc = r + dr, c + dc
            if 0 <= nr < n and 0 <= nc < m and grid[nr][nc] == 0:
                new_pos = (nr, nc)
                if new_pos not in visited:
                    visited.append(new_pos)
                    queue.append((new_pos, steps + 1))
    return -1

def greedy_best_first_search(grid):
    n, m = len(grid), len(grid[0])
    start = (0, 0)
    goal = (n - 1, m - 1)

    open_list = [(start, [start])]
    visited = []

    while open_list:
        best_index = 0
        best_h = 999999
        for i in range(len(open_list)):
            h = manhattan(open_list[i][0], goal)
            if h < best_h:
                best_h = h
                best_index = i

        pos, path = open_list.pop(best_index)
        visited.append(pos)

        if pos == goal:
            return (path, len(path) - 1)

        r, c = pos
        for dr, dc in [(1,0), (-1,0), (0,1), (0,-1)]:
            nr, nc = r + dr, c + dc
            if 0 <= nr < n and 0 <= nc < m and grid[nr][nc] == 0:
                new_pos = (nr, nc)
                if new_pos not in visited:
                    open_list.append((new_pos, path + [new_pos]))

    return (None, -1)

n = int(input("Enter rows: "))
m = int(input("Enter columns: "))
grid = []
for i in range(n):
    row = list(map(int, input(f"Enter row {i}: ").split()))
    grid.append(row)

gbfs_path, gbfs_steps = greedy_best_first_search(grid)
bfs_steps = bfs_shortest_steps(grid)

print("\nGBFS Path:", gbfs_path)
print("GBFS Steps:", gbfs_steps)
print("BFS Shortest Steps:", bfs_steps)

if gbfs_steps == bfs_steps:
    print("GBFS found the shortest path.")
else:
    print("GBFS did NOT find the shortest path.")
```
